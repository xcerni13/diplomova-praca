### Vysvetlenie pojmov

Algoritmus je konečná množina pravidiel, ktorá popisuje postupnosť operácií na riešenie určitého problému.
<br>
<br>
 Inak povedané, je to presný popis krokov k vyriešeniu úlohy. Tento popis môže byť zapísaný rôznymi spôsobmi, avšak dôležité je to, že je možné podľa neho postupovať a prepísať do ľubovoľnej formy a vždy bude možné vyriešiť zadanú úlohu.

<div class="alert alert-block alert-info">
<b>Algoritmus</b> je usporiadaná sada pravidiel, ktoré jednoznačne vyriešia zadanú úlohu. 

</div>

Typický príklad algoritmu, s ktorým sa môžete stretnúť je riešenie Rubikovej kocky.

Pre zápis algoritmu existuje množstvo spôsobov. Najdôležitejšie je, aby bol zapísaný jasne a bolo možné podľa tohto zápisu algoritmus vykonať. Typické spôsoby zápisu algoritmu sú:
- prirodzený jazyk
- pseudokód
- vývojový diagram (flowchart)
- programovací jazyk

##### Algoritmický jazyk - pseudokód

Ide o spôsob zápisu algoritmu, ktorý je medzi prirodzeným jazykom a programovacím jazykom. Používa programovacie konštrukcie ako cyklus a podmienka, ale zapisuje ich prirodzeným jazykom. Oproti prirodzenému jazyku je presnejší.

##### Vývojový diagram - flowchart

Flowchart je grafické znázornenie algoritmu. Na obrázku nižšie môžete vidieť jednotlivé grafické prvky, ktoré sa v ňom používajú. Každý prvok ma priradené použitie.

Hlavnou motiváciou pre písanie algoritmov, je snaha tvoriť efektívne riešenia problémov. Takmer výhradne ide o problémy, ktoré sa opakujú a preto potrebujeme riešenie, ktoré ich vždy vyrieši, aj keď sa niektoré detaily zmenia. Aj keď v prvom príklade nebol riešený programovací problém, môžete na ňom vidieť, že nielen v programovaní sa dajú algoritmy využiť. Typické problémy, na ktoré je potrebné nájsť algoritmické riešenie sú programovacie.

### Pokračovanie úlohy

Pri vytváraní algoritmu je dôležité vyriešiť problém čo najviac všeobecne. To znamená hľadať riešenie na širšiu skupinu problémov, nielen jeden kontrétny. Všeobecnosť je jedna zo základných vlastností algoritmu.

<div class="alert alert-block alert-success">

### Vysvetlenie pojmov

Vlastnosti algoritmov:
- deterministický - jednoznačné kroky
- elementárny
- konečný - ukončí sa
- rezultatívny - po ukončení dá výsledok
- všeobecný - nerieši iba konkrétny prípad, ale celú skupinu podobných problémov
