### Vysvetlenie pojmov

<div class="alert alert-block alert-info">
for cyklus je riadiaca štruktúra, ktorá opakuje príkazy, ktoré sú v tele cyklu zapísané
</div>

Často sa môžete stretnúť so situáciou, kedy potrebujete opakovať nejaký príkaz niekoľkokrát. Namiesto kopírovania toho istého príkazu toľkokrát, koľkokrát ho potrebujeme vykonať, existuje cyklus.
<br>
Na tento účel je možné využiť <b>for cyklus </b>. 

##### Syntax

```python
for premenná in range(počiatočná hodnota, konečná hodnota):
    telo cyklu
```

<div class="alert alert-block alert-info">
<b>for</b> - kľúčové slovo pre začiatok cyklu
<br>
<b>premenná</b> - riadiaca premenná, pomocou ktorej sa počítajú jednotlivé opakovania, nadobúda hodnoty z rozsahu určeného v zátvorkách
<br>
<b>in range</b> - v akom rozsahu budú opakovania - aké hodnoty bude naberať riadiaca premenná
<br>
<b>(počiatočná hodnota, konečná hodnota)</b> -počet opakovaní - začiatok je štandardne na čísle 0, v zátvorkách je možné napísať jedno číslo, ktoré určuje počet opakovaní, alebo začiatočnú a konečnú hodnotu riadiacej premennej.
<br>
<b>telo cyklu</b> - príkazy, ktoré budú vykonané v každom opakovaní cyklu
</div>

Na začiatku for cyklu prebieha inicializácia riadiacej premennej. Následne prebieha kontrola podmienky na začiatku cyklu. Po kontrole podmienky sú spustené príkazy v tele cyklu a nakoniec je inkrementovaná (zväčšená) riadiaca premenná.

```python
for i in range(0, 5):
    print(i)
```

<b>Riadiaca premenná</b> môže mať ľubovoľný identifikátor. Typicky sa využívaju písmená. Zvyčajne vidíme <b>i</b>, ako iterátor. Premenná je platná v bloku cyklu a môže byť použitá aj na iné účely v cykle. Nie je potrebné meniť jej hodnotu v každom opakovaní, Po každom opakovaní cyklu je hodnota inkrementovaná.
<br>
Jedno opakovanie nazývame <b>iterácia</b>.
<br>
<br>
<b>Počet opakovaní</b> je možné definovať jedinou hodnotou n. Vtedy cyklus začne opakovať na hodnote 0 a skončí na hodnote n-1. 

```python
for i in range(5):
    print("hello world")
    
```

Prípadne je možné definovať počiatočnú a konečnú hodnotu riadiacej premennej.

```python
for k in range(0, 5):
    print("hello world")
```

Cyklus je možné využiť na ľubovoľnú úlohu, keď potrebujeme nejaký príkaz vykonať opakovane.

```python
for i in range(0, 5):
    print(i)
```

Podobne ako pri podmienkach, aj pri for cykle vzniká nový <b>blok kódu</b>. Z toho vyplýva, že všetky príkazy, ktoré majú byť vykonané v tele cyklu, musia byť odsadené. Takisto to znamená, že premenné definované v cykle sú platné iba v tomto bloku.

##### Vnorený cyklus

Vnorený cyklus je pojem, ktorý hovorí o situácii, keď sa v tele cyklu nachádza ďalší cyklus. 

Predstavte si postup z úvodnej úlohy. Každý riadok ktorý postava prechádzala bolo potrebné prejsť po jednom poli. V tejto situácii je ideálne použiť vnorený cyklus. Postup zostáva rovnaký ako v pôvodnom riešení. Pomocou cyklu prejdeme všetky riadky. V každom opakovaní prejdeme všetky polia v danom riadku. Ale na prejdenie riadku použijeme ďalší cyklus, ktorý prejde všetky polia v danom riadku.

```python
size = 5

for row in range(size):
    for square in range(size):
        print("x", end ='')
    print("\n")
```
